```python
from pathlib import Path
import pandas as pd
```

```python
# Store filepath in a variable
file = Path("Resources/user_behavior_dataset.csv")
```

```python
# Read our data file with the Pandas library
original_df = pd.read_csv(file)
original_df.head()
```

```python
original_df.columns
```

```python
columns_list = original_df.columns.tolist() 
columns_list
```

```python
labels = [col.replace(' ', '_') for col in original_df.columns]

labels_lower = [item.lower() for item in labels]

labels_lower
```

```python
columns_dict = dict(zip(columns_list, labels_lower))
columns_dict
```

```python
columns_dict_new = {'User ID': 'user_id',
 'Device Model': 'device_model',
 'Operating System': 'operating_system',
 'App Usage Time (min/day)': 'app_usage_minutes_per_day',
 'Screen On Time (hours/day)': 'screen_time_hours_per_day',
 'Battery Drain (mAh/day)': 'battery_drain_mah_per_day',
 'Number of Apps Installed': 'number_of_apps_installed',
 'Data Usage (MB/day)': 'data_usage_mb_per_day',
 'Age': 'age',
 'Gender': 'gender',
 'User Behavior Class': 'user_behavior_class'}
```

```python
df_columns_final = original_df.rename(columns=columns_dict_new)
df_columns_final
```

```python
df_columns_final.to_csv('data.csv', index=False)
```
